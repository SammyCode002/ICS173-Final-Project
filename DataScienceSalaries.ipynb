```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



```

```python
# Load the CSV file into a DataFrame
csv_file_path = 'ds_salaries.csv'  # Update the path if needed
data = pd.read_csv(csv_file_path)
data.head()

```

```python
# Distribution of Salaries
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.histplot(data['salary_in_usd'], bins=30, kde=True)
plt.title('Distribution of Salaries (in USD)')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.show()
```

```python
# Average Salaries by Experience Level
plt.figure(figsize=(10, 6))
average_salary_by_experience = data.groupby('experience_level')['salary_in_usd'].mean().sort_values()
sns.barplot(x=average_salary_by_experience.index, y=average_salary_by_experience.values, color="skyblue")
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary in USD')
plt.show()

```

```python
# Remote Ratio and Salary Comparison
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='remote_ratio', y='salary_in_usd', hue='experience_level', palette="deep", alpha=0.6)
plt.title('Salary in USD vs. Remote Ratio')
plt.xlabel('Remote Ratio')
plt.ylabel('Salary in USD')
plt.legend(title='Experience Level')
plt.show()
```

```python
# Job Count by Company Size
plt.figure(figsize=(10, 6))
job_count_by_company_size = data['company_size'].value_counts().sort_index()
sns.barplot(x=job_count_by_company_size.index, y=job_count_by_company_size.values, color="teal")
plt.title('Number of Jobs by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Number of Jobs')
plt.show()
```
